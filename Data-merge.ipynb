```python
import glob
import pandas as pd
```

```python
path = "C:/Users/eceak/Apple Health Project/data"
```

```python
file_list = glob.glob(path + "/*.xlsx")
```

```python
excl_list = []
```

```python
for file in file_list:
    excl_list.append(pd.read_excel(file))
```

```python
excl_list
```

```python
excl_merged = pd.concat(excl_list, ignore_index=True)
excl_merged.to_excel('apple_watch_data.xlsx', index=False)
```
